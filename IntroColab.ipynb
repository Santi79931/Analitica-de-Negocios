0.Cargar las librerias o base de datos.

```python
import numpy as np #Es para las funciones numericas para transformacion y analisis de datos.
import pandas as pd #Es para conectar el sistema con la base de datos.
```

1.Conexion con la base de datos

```python
nxl="/content/Quiz1.xlsx"
XDB=pd.read_excel(nxl, sheet_name=0) #Dataframe: tabla de datos - Tablas de Word
XDB.head()
```

2. Llamar la variable de caracterizacion (Edad)

```python
XDe=XDB[['Edad']]
print(XDe)
XDe=XDe.dropna() #Sirve para evitar las celdas vacias.
XDe=np.array(XDe)
```

3. Se calcula la media y la desviacion estandar de la edad

```python
print("La media de la edad es:", np.mean(XDe))
print("La desviacion estandar de la edad es:", np.std(XDe))
print("La edad minima es:", np.mean(XDe)-np.std(XDe))
print("La edad maxima es:", np.mean(XDe)+np.std(XDe))
```

4. Histograma de Frecuencias - Agrupar los datos por intervalos

```python
import matplotlib.pyplot as plt
plt.hist(XDe, bins=15, color='darkgreen')
plt.show()
```

5. Calcular el valor de pertenencia para una persona de 60 años (Este valor esta entre 0 y 1) - Afinidad de un valor por una variable aleatoria.

```python
VP=np.exp(-0.5*((60-np.mean(XDe))/np.std(XDe))**2)
print("El valor de pertenencia de una persona de 60 años es:", VP)

VP=np.exp(-0.5*((50-np.mean(XDe))/np.std(XDe))**2)
print("El valor de pertenencia de una persona de 50 años es:", VP)

VP=np.exp(-0.5*((49.12-np.mean(XDe))/np.std(XDe))**2)
print("El valor de pertenencia de una persona de 49.12 años es:", VP)
```
