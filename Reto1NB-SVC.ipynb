**Modelo Naibe Bayes**
Analisis de la base de datos Longitudinal Survey usando el modelo Naive Bayes.
0. Carga de las librerias de trabajo.

```python
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import confusion_matrix
```

1. Carga de la base de datos

```python
nxl='/content/1. BD1_Longitudinal Survey_Int.xlsx'
XDB=pd.read_excel(nxl, sheet_name=0)
XDB.dropna(inplace=True) #Eliminar celdas vacias

XDB=XDB[['Age','Mother_Edu','Father_Edu','FamilySize','Self_Esteem','Height','Weight','Income', 'Urban']]
#Datos de Entrada
XD=np.array(XDB[['Age','Mother_Edu','Father_Edu','FamilySize','Self_Esteem','Height','Weight','Income']])
#Datos de Salida - Valor deseado - Lo que quiero que aprenda el modelo
yd=np.array(XDB['Urban'])
XDB
```

2. Implemetar el modelo

```python
mnb=GaussianNB()
mnb.fit(XD,yd)
ydp=mnb.predict(XD)

#Parametros del modelo por categoria - Urbana o Rural
u=mnb.theta_
sigma=np.sqrt(mnb.var_)

#Metrica de evaluación - Como se comporto el modelo identificando Urbano o Rural
cm=confusion_matrix(yd,ydp)
print("La matriz de confusion es:")
print(cm)
```

3. Evaluacion del modelo

```python
VN=cm[0,0] #Verdaderos Negativos
FP=cm[0,1] #Falsos Positivos
FN=cm[1,0] #Falsos Negativos
VP=cm[1,1] #Verdaderos Positivos

Exactitud=(VP+VN)/len(XD)
print("La exactitud del modelo es:", Exactitud) #Comportamiento General

TE=(FN+FP)/len(XD)
print("La tasa de error del modelo es:", TE) #Porcentaje de Equivocaciones

Sensibilidad=VP/(VP+FP)
print("La sensibilidad del modelo es:", Sensibilidad) #Como se comporto frente a los positivos

Especificidad=VN/(VN+FN)
print("La especificidad del modelo es:", Especificidad) #Como se comporto frente a los negativos

Precision=VP/(VP+FN)
print("La precición del modelo es:", Precision) #Como le pego a los clasificados positivos
```

4. Ejemplo de una persona especifica con caracteristicas dadas

```python
Xun=np.array([[34,9,6,3,30,65,135,50000]]) #Persona especifica
ydun=mnb.predict(Xun)
print(ydun)
```
