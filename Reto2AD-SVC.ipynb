```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt #Graficador
from sklearn.tree import DecisionTreeClassifier

from google.colab import drive
drive.mount('/content/drive')
```

1.Cargar los datos

```python
nxl='/content/drive/MyDrive/Analitica de Negocios/2. BD1_Travel_Data_int.xlsx'
XDB=pd.read_excel(nxl,sheet_name=0)
XDB=XDB.dropna() #Borra las celdas vacias
XDB.head(100)

```

```python
#Seleccion de las variables a trabajar
XD=np.array(XDB[['College','FoodSpend','Income','TravelSpend']]) #Datos de Entrada
print(XD)
yd=np.array(XDB[['CreditCard']]) #Datos de Salida o datos que el modelo va a pronosticar
print(yd)
```

2. Creación del Modelo del arbol de decisión

```python
mar=DecisionTreeClassifier(criterion='gini', max_depth=4) #4 variables de entrada
mar.fit(XD,yd)
ydp=mar.predict(XD)

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(yd,ydp)
print(cm)

#Evaluación del modelo
VN=cm[0,0]; FP=cm[0,1]; FN=cm[1,0]; VP=cm[1,1]

Ex=(VN+VP)/(VN+VP+FN+FP)  #Que porcentaje de los datos pronostico correctamente
print("La exactitud alcanzada por el modelo es: ",Ex)

Se=VP/(FN+VP) #Como le fue con los positivos
print("La sensibilidad alcanzada por el modelo es: ",Se)

Es=VN/(VN+FP) #Como le fue con los negativos
print("La especificidad alcanzada por el modelo es: ",Es)
```

3. Graficación del arbol

```python
#Librerias para gráficar el arbol
from sklearn.tree import export_graphviz
from pydotplus import graph_from_dot_data

vs=['College','FoodSpend','Income','TravelSpend'] #Títulos de los nodos del Arbol
dot_graph=export_graphviz(mar,feature_names=vs)
graph=graph_from_dot_data(dot_graph)
graph.write_png('arbolReto.png')
```

4.Pronostico de los personajes especificos/Nuevos solicitantes

```python
XDB2=pd.read_excel(nxl,sheet_name=1)
XDB2=XDB2.dropna() #Borra las celdas vacias
XDB2.head()
```

```python
XD2= np.array(XDB2[['College','FoodSpend','Income','TravelSpend']]) #Datos de entrada
print(XD2)

#Pronosticos
ydp2=mar.predict(XD2)
print(ydp2)
print("El número de clientes que usaron la tarjeta es:" ,np.sum(ydp2))
print("El numero total de Datos es:" , len(ydp2))
print("El numero de clientes que no usaron la tarjeta es:" , len(ydp2)-np.sum(ydp2))
```
