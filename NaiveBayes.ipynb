**Modelo de Preaprobacion -Naibe Bayes**
Base de datos de Solicitantes de Credito mediante el modelo Naive Bayes.
0. se procede con la carga de las librerias de trabajo.

```python
import numpy as np #Funciones Matematicas.
import pandas as pd #Comunicacion con excel.
from sklearn.naive_bayes import GaussianNB #Modelo de Clasificacion - Funcion Gaussiana.
from sklearn.metrics import confusion_matrix #Matriz de Confusion - La metrica de Medida.
```

1. Secargan los dato.


```python
nxl = '/content/SolicitantesCreditoNaiveBayes.xlsx'
XDB=pd.read_excel(nxl, sheet_name=2) #Traer los datos de la hoja AAA
XDB.dropna() #Eliminar datos faltantes, celdas vacias

#Seleccionamos las variables de interes.
XDB=XDB[['Edad','Ingresos','Egresos','Cuota (COP)','PreApr']]

#Datos de Entrada
XD=np.array(XDB[['Edad','Ingresos','Egresos','Cuota (COP)']])
#Datos de Salida - Valor deseado - Lo que quiero que aprenda el modelo
yd=np.array(XDB['PreApr'])
```

2. Implementacion del Modelo

```python
mnb=GaussianNB() #Llamamos el modelo
mnb.fit(XD,yd) #Entrenamos el modelo
ydp=mnb.predict(XD) #Muestra los datos de salida que el obtuvo

#Parametros del modelo por categoria - PreApr/PreNeg
u=mnb.theta_ #Se estiman las medidas
sigma=np.sqrt(mnb.var_) #Se estiman las desviaciones

#Las vaariaciones de las variables
LI=u-sigma
LS=u+sigma

print("La media de las variables por categoria: ")
print(u)
print("El limite inferior de la variación de las variables por categoria es:")
print(LI)
print("El limite superior de la variación de las variables por categoria es:")
print(LS)

#Metrica de evaluación - Como se comporto el modelo identificando la PreApr-PreNeg
cm=confusion_matrix(yd,ydp)
print("La matriz de confusion es:")
print(cm)
```

3. Se procede con la evaluacion del modelo.


```python
VN=cm[0,0] #Verdaderos Negativos
FP=cm[0,1] #Falsos Positivos
FN=cm[1,0] #Falsos Negativos
VP=cm[1,1] #Verdaderos Positivos

Exactitud=(VP+VN)/len(XD)
print("La exactitud del modelo es:", Exactitud) #Comportamiento General

TE=(FN+FP)/len(XD)
print("La tasa de error del modelo es:", TE) #Porcentaje de Equivocaciones

Sensibilidad=VP/(VP+FP)
print("La sensibilidad del modelo es:", Sensibilidad) #Como se comporto frente a los positivos

Especificidad=VN/(VN+FN)
print("La especificidad del modelo es:", Especificidad) #Como se comporto frente a los negativos

Precision=VP/(VP+FN)
print("La precición del modelo es:", Precision) #Como le pego a los clasificados positivos

PredNeg=VN/(VN+FP)
print("La probabilidad de Negativo del modelo es:", PredNeg) #Como le pego a los clasificados negativos

```

4. Que pasa con una sola persona?

```python
Xun=np.array([[38,1356,1686,18]]) #Persona especifica del ejemplo de excel
ydun=mnb.predict(Xun) #
print(ydun)
```
